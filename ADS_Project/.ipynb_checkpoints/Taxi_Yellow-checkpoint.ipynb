```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
df = pd.read_csv("D://Taxi Data//yellow_tripdata_2018-01.csv")
```

```python
df = df.loc[:,["tpep_pickup_datetime","PULocationID"]]
df['WEEKEND'] = ((pd.DatetimeIndex(df.tpep_pickup_datetime).dayofweek) >= 5)
df = df[df['WEEKEND'] == True]
df['tpep_pickup_datetime'] = df['tpep_pickup_datetime'].map(lambda x: x.split(" ")[0])
df.head()
```

```python
df = df.groupby(["tpep_pickup_datetime","PULocationID"]).sum()
df.reset_index(inplace=True)
df.head()
```

```python
df2 = gpd.read_file("D:\\Taxi Data\\taxi_zones.shp")
df2.head()
```

```python
df3 = pd.merge(df,df2,left_on="PULocationID",right_on="LocationID",how="left")
df3.head()
```

```python
df3.to_csv("yellow_tripdata_2018-01_clean.csv")
```
