```python
import pandas as pd
from sklearn.metrics import log_loss
import numpy as np

data_dir1 = '../input/NCAA/'
data_dir2 = '../working/'

actual_pred = pd.read_csv(data_dir1+'NCAATourneyDetailedResults.csv')#actual results file

actual_pred.head()
```

```python
actual_pred = actual_pred[['Season','WTeamID','LTeamID']]
actual_pred = actual_pred.loc[actual_pred['Season']>=2014,:]
actual_pred.head()
```

```python
pred_pred = pd.read_csv(data_dir2+'NCAA_short4.csv')#predicted results file
pred_pred.head()
```

```python
def get_year(ID):
    return int(ID.split('_')[0])

def get_t1(ID):
    return int(ID.split('_')[1])

def get_t2(ID):
    return int(ID.split('_')[2])

pred_pred['Season'] = pred_pred['ID'].apply(get_year)
pred_pred['WTeamID'] = pred_pred['ID'].apply(get_t1)
pred_pred['LTeamID'] = pred_pred['ID'].apply(get_t2)
pred_pred.drop(['ID'],axis=1,inplace=True)
pred_pred.head()
```

```python
actual_pred = pd.merge(left=actual_pred,right=pred_pred,how='left',on=['Season','WTeamID','LTeamID'])
actual_pred.rename(columns={'Pred':'Pred1'},inplace=True)
actual_pred['Pred1'].fillna(0,inplace=True)

actual_pred = pd.merge(left=actual_pred,right=pred_pred,how='left',left_on=['Season','WTeamID','LTeamID'],right_on=['Season','LTeamID','WTeamID'])
actual_pred.rename(columns={'Pred':'Pred2'},inplace=True)
actual_pred['Pred2'].fillna(0,inplace=True)
actual_pred['Pred3'] = 1 - actual_pred['Pred2']
actual_pred['Pred3'] = actual_pred['Pred3'].apply(lambda x: x if x<1 else 0)
actual_pred.head()
```

```python
actual_pred['Pred4'] = actual_pred['Pred1'] + actual_pred['Pred3']
actual_pred.drop(['WTeamID_y','LTeamID_y','Pred1','Pred2','Pred3'],inplace=True,axis=1)
actual_pred.rename(columns={'WTeamID_x':'WTeamID','LTeamID_x':'LTeamID'},inplace=True)
actual_pred['Result'] = 1
actual_pred.head()
```

```python
log_loss(np.array(actual_pred['Result'].values),np.array(actual_pred['Pred4'].values),labels=[1,0])
```
